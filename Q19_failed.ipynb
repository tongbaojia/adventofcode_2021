```python
class Raw_measure:
    def __init__(self, x, y, z, parent, uid):
        self.x = x
        self.y = y
        self.z = z
        self.parents = set([parent])
        self.uid = uid
        self.distances = dict()
        self.permute_dist = dict()
        self.isordered_dist_keys = None
        self.cousin = set([uid])
        self.origin = [0, 0, 0]

    def dist(self, another_measure):
        """Return a tuple of 3 differences"""
        self.distances[
            (
                self.x - another_measure.x,
                self.y - another_measure.y,
                self.z - another_measure.z,
            )
        ] = another_measure.uid
        another_measure.distances[
            (
                -self.x + another_measure.x,
                -self.y + another_measure.y,
                -self.z + another_measure.z,
            )
        ] = self.uid

    def __str__(self):
        return f"{self.x},{self.y},{self.z}, {self.parents}, {self.uid}"

    #     def permuate_dist(self):
    #         n_permute = 0
    #         for a, b, c in itertools.permuations([self.x, self.y, self.z]):
    #             for i in range(4):
    #                 flip_index = i
    #                 self.permute_dist[n_permute] = (
    #                     (-1) * a if i == 0 else a,
    #                     (-1) * b if i == 1 else b,
    #                     (-1) * c if i == 2 else c,
    #                 )
    #                 n_permute += 1

    def distance_disordered_keys(self):
        if self.isordered_dist_keys is None:
            isordered_dist_keys = []
            for k in self.distances.keys():
                x, y, z = k
                isordered_dist_keys.append(set([abs(x), abs(y), abs(z)]))
            self.isordered_dist_keys = isordered_dist_keys
        return self.isordered_dist_keys

    def n_intersection(self, another_measure):
        ans = 0
        for i in self.distance_disordered_keys():
            if i in another_measure.distance_disordered_keys():
                ans += 1
        # aha, shared, mutate
        if ans == 11:
            self.parents = self.parents.union(another_measure.parents)
            another_measure.parents = self.parents.union(another_measure.parents)
            self.cousin.add(another_measure.uid)
            another_measure.cousin.add(self.uid)
            
        return ans
```
